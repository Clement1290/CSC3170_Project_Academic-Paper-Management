```python
import pymysql
import pandas as pd
```

```python
passward = 'YOUR DATABASE PASSWARD'
```

```python
#author
Author = pd.read_csv('Author.csv')
con = pymysql.connect(user="root",
                      passwd=passward,
                      db="GoogleScholar",
                      host="127.0.0.1",
                      charset='utf8mb4')
cur = con.cursor()

# 4. 插入数据
for index, row in Author.iterrows():
    insert_query = '''
    INSERT INTO author (author_link, author_name, h_index)
    VALUES (%s, %s, %s);
    '''

    cur.execute(insert_query, (row['author_link'], row['author_name'], row['author_h_index']))

# 5. 提交更改并关闭连接
con.commit()
cur.close()
cur.close()
```
